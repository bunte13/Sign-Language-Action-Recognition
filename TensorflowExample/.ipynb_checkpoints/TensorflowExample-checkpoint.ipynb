```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Churn.csv')
```

```python
X = pd.get_dummies(df.drop(['Churn','Customer ID'],axis = 1))
y = df['Churn'].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

# 1. Import Dependencies

```python
#Sequential: A linear stack of layers in Keras. It allows you to build a model layer by layer.
#load_model: A function to load a pre-trained Keras model.
from tensorflow.keras.models import Sequential,load_model
#Dense: A fully connected layer in a neural network.
from tensorflow.keras.layers import Dense
#accuracy_score: A function from scikit-learn to calculate the accuracy of predictions.
from sklearn.metrics import accuracy_score
```

 # 2. Bild and compile model

```python
model = Sequential()
model.add(Dense(units = 32 ,activation = 'relu' , input_dim = len(X_train.columns)))
model.add(Dense(units = 64,activation = 'relu'))
model.add(Dense(units = 1,activation = 'sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

# 3. Fit, predict, and evaluate

```python
model.fit(X_train, y_train, epochs = 200, batch_size = 32) #training our model
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test,y_hat)
```

# 4. Saving and reloading

```python
model.save('tfmodel_example.h5')
```

```python
del model
```
